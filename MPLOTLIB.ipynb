## Walking through the Matplotlib pipeline

#### Importing the module

```
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
# now lets create a custom data to plot
cust_data_x = [i for i in range(20)]
cust_data_y = [2 * i for i in range(20)]
print(cust_data_x)
print(cust_data_y)
```

#### Q. How do you expect the line plot to look like?
- What kind of plot shall we use to visualize the points ?

```
plt.plot(cust_data_x, cust_data_y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```
# creating another kind of data
cust_data_x = [i for i in range(-10, 11, 1)]
cust_data_y = [i ** 2 for i in cust_data_x]
print(cust_data_x)
print(cust_data_y)
```

#### Q. Answer the same questions as before!

```
x = np.arange(1,10,1)
y = np.arange(1,10,1)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y)
plt.show()
```

```
plt.plot(cust_data_x, np.sin(cust_data_y))
plt.show()
```

```
cust_data_x_sin = np.linspace(-10,10,100)
cust_data_y_sin = np.sin(cust_data_x_sin)
```

```
# creating the figure and visualizing it
fig = plt.figure(figsize=(16,8))
plt.plot(cust_data_x_sin, cust_data_y_sin)
plt.show()
```

```
# drawing in the same plot
fig = plt.figure(figsize=(12,8))
plt.plot(cust_data_x, cust_data_y, c='r', linestyle='-') # 'c' property for color
plt.plot(cust_data_x_sin, cust_data_y_sin, c='b')
plt.show()
```

# Creating Subplots

```
# lets plot the above two plots side by side
plt.figure(figsize=(16,7))
plt.subplot(1, 2, 1)
plt.plot(cust_data_x, cust_data_y, c='r')
plt.subplot(1, 2, 2)
plt.plot(cust_data_x_sin, cust_data_y_sin, c='b')
plt.show()
```

Q. Create 4 subplots in a 2 by 2 grid, in each cell, use different linestyle or properties for the same data that you wish to use.

```
x=np.arange(-3.14,3.15,.1)
y1=np.sin(x)
y2=np.cos(x)
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.plot(x,y1, c='r')
plt.plot(x,y2, c='b')
plt.show()
y3=y1*y2
plt.subplot(2,2,3)
plt.plot(x,y3,c='g')
y4=y1*y1
plt.subplot(2,2,4)
plt.plot(x,y4,c='y')
plt.show()
```
